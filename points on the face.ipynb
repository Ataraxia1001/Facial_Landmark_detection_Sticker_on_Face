```python
import cv2
import numpy as np
```

```python
img = cv2.imread('./01_Indoor/indoor_013.png') # default flag: cv2.IMREAD_COLOR
#points = cv2.imread('./01_Indoor/indoor_003.pts')
```

```python
img.shape
# all images have different size
```

```python
h, w, _ = img.shape
h, w, _
```

```python
f = open("./01_Indoor/indoor_013.pts", "r")
f2 = f.read() # convert _io.TextIOWrapper file to string
type(f2)
```

```python
l = len(f2)
l
```

```python
print(f2)
```

```python
f3 = f2.split()
f3
```

```python
f4 = f3[5:-1]
f4
```

```python
land_list = []
for i in range(0, len(f4), 2):
    x = float(f4[i])
    y = float(f4[i+1])
    x = x/w
    y = y/h
    land_list.append(x)
    land_list.append(y)
print(land_list)
```

```python
land_arr = np.array(land_list, dtype=np.float32)
land_arr
```

```python
for i in range(0, len(land_arr), 2):
    x = int(land_arr[i]*w)
    y = int(land_arr[i+1]*h)
    #print(x,y)
    img = cv2.circle(img,(x, y), 3, (0,255,0), -1)
    num = int(i/2 + 1)
    num_text = str(num)
    text_origin = (x,y+10)
    TEXT_FACE = cv2.FONT_HERSHEY_DUPLEX
    TEXT_SCALE = 0.3
    TEXT_THICKNESS = 1
    img = cv2.putText(img, num_text, text_origin, TEXT_FACE, TEXT_SCALE, (0,0,255), TEXT_THICKNESS, cv2.LINE_AA)
```

```python
# Using cv2.imshow() method
# Displaying the image
# img = cv2.resize(img, (960, 540))   
cv2.imshow('face', img)
  
# waits for user to press any key
# (this is necessary to avoid Python kernel form crashing)
cv2.waitKey(0)
  
# closing all open windows
cv2.destroyAllWindows()
```
